```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams["font.size"] = 18
mpl.rcParams['mathtext.fontset'] = 'stix'
mpl.rcParams['font.family'] = 'STIXGeneral'
```

```python
praefix = ["5", "10", "15", "10_100ep"]
dat = list()
for i,p in enumerate(praefix):
    file_name = f"training_noPooling/test_al{p}.txt"
    dat_ = np.loadtxt(file_name, delimiter=",")
    dat.append(dat_)
```

```python
fig = plt.figure()
ax_acc = fig.add_subplot()
ax_acc.plot(dat[3][:, 1], c="g", label="Accuracy")
ax_acc.set_xlabel("Epoch")
ax_acc.set_ylabel("Accuracy")
ax_loss = ax_acc.twinx()
ax_loss.set_ylabel("Test Loss")
ax_loss.plot(dat[3][:,0], c="orange", label="Test Loss")
fig.legend(ncol=2, loc="upper center", bbox_to_anchor=(0.51, 1.05))
fig.savefig("percNoPoolling.png", dpi=300, bbox_inches="tight")
```

```python
dat = np.loadtxt("training_pooling/test_al10_1topk_100ep.txt", delimiter=",")

fig = plt.figure()
ax_acc = fig.add_subplot()
ax_acc.plot(dat[:, 1], c="g", label="Accuracy")
ax_acc.set_xlabel("Epoch")
ax_acc.set_ylabel("Accuracy")
ax_loss = ax_acc.twinx()
ax_loss.set_ylabel("Test Loss")
ax_loss.plot(dat[:,0], c="orange", label="Test Loss")
fig.legend(ncol=2, loc="upper center", bbox_to_anchor=(0.51, 1.05))
fig.savefig("percPoolling.png", dpi=300, bbox_inches="tight")
```
